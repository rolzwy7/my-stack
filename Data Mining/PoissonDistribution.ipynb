    <h1>Poisson Distribution</h1>

One nanogram of Pluton-239 will have an average of 2.3 radioactive decays per second, and the number of decays will follow Poisson distribution.

<b>What is probability that in a 2 seconds period there are exactly 3 radioactive decays?</b>

<b>X</b> - number of decays over 2 seconds period.

```python
from math import exp, factorial
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

def poisson(x, lambda_):
    return (lambda_**x * exp(-lambda_))/factorial(x)

rd_per_sec = 2.3

# P(X = x) , P(X = 3), x=3
lambda_ = 2 * rd_per_sec
x = 3

print("P(X=3) = %.2f%%" % (poisson(x, lambda_)*100))
```

```python
x = list(range(0, 10+1))
y = list(map(lambda a: poisson(a, lambda_), x))
df = pd.DataFrame(columns=["x", "y"])
df["x"] = x
df["y"] = y
```

```python
plt.figure(figsize=(15,7))
plot = sns.barplot(x="x", y="y", data=df)
plt.grid()
```
