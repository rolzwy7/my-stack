```python
%matplotlib inline

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import rc as matplotlib_rc
import os
import math

import matplotlib.ticker as plticker

rok_locator = plticker.MultipleLocator(base=1.0)

matplotlib_rc('font', **{'sans-serif' : 'Arial',
                         'family' : 'sans-serif'})
```

```python
DATA_DIR = os.path.join("..", "_data")

memory_df = pd.read_csv(os.path.join(DATA_DIR, "io_lab_2_memory.csv"), sep=",")
time_df = pd.read_csv(os.path.join(DATA_DIR, "io_lab_2_time.csv"), sep=",")


time_df["Pomiar"] = time_df["Pomiar"].apply(lambda x: math.log10(x))
```

```python
plt.figure(figsize=(15, 7))
plt.title('Zużycie pamięci w funkcji rozmiaru kraty')

chart_1 = sns.lineplot(x="Rozmiar kraty", y="Pamięć [MB]", markers=True, data=memory_df)
#chart_1.xaxis.set_major_locator(rok_locator)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.title('Zużycie pamięci w funkcji rozmiaru kraty')

chart_2 = sns.lineplot(x="Rozmiar kraty", y="Pomiar", hue="Rodzaj", style="Rodzaj", markers=True, data=time_df)
#chart_1.xaxis.set_major_locator(rok_locator)
plt.grid()
plt.show()
```
